```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
link= 'https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv'
code = '?raw=true'
url = link+code
df = pd.read_csv(url,sep=',',index_col=1)
```

```python
cuantos=df.groupby('Country').size().reset_index(name='Conteo')
cuantos
```

```python
plt.pie(cuantos['Conteo'],labels=cuantos['Country'],autopct='%0.1f %%')
plt.axis("equal")
plt.title("Distribucion del ebola por paises")
plt.savefig("Pastel_Paises.png",dpi=300,format='png')
plt.show()
```

```python
SL=df[df['Country']=="Sierra Leone"]
```

```python
df['Country']=="Sierra Leone"
```

```python
mean_values=SL['No. of confirmed deaths'].mean()
mean_values
```

```python
SL.fillna(mean_values,inplace=True)
```

```python
Dates=SL.index[0:-1:10]
NCC=SL['No. of confirmed cases'][0:-1:10]
NCD=SL['No. of confirmed deaths'][0:-1:10]
```

```python
fig=plt.figure(figsize=(13,7))
plt.plot(Dates,NCC,"o--y",ms=2,label="Infectados")
plt.bar(Dates,NCC,color='yellow',alpha=0.4)
plt.plot(Dates,NCD,'o--g',ms=4,label='Decesos')
plt.bar(Dates,NCD,color='green',alpha=0.3)
plt.grid(color="gray",linewidth="2",ms=20)
plt.title("Avance del Ã©bola en Sierra Leona")
plt.ylabel("Casos confirmados")
plt.xlabel("Periodo de tiempo")
plt.xticks(rotation=90,fontsize=7)#Fechas rotadas y tamaÃ±o de las fechas.
plt.legend()
plt.savefig("infectadosydecesos.png",dpi=300,format='png')
plt.show()
```

```python
PIvsD=100*SL['No. of confirmed deaths']/SL['No. of confirmed cases']
SL['PIvsD']=PIvsD
```

```python
recuperados=((100-PIvsD)/100)*SL['No. of confirmed cases']
SL['recuperados']=recuperados
```

```python
SL['PR']=100-PIvsD
```

```python
SL.head()
```

```python
PIvsD_mean=SL['PIvsD'].mean()
PR_mean=SL['PR'].mean()
porcentajes=[PIvsD_mean,PR_mean]
etiquetas=['PIvsS','PR']
```

```python
plt.pie(porcentajes, labels=etiquetas, autopct='%0.1f %%', colors=['green','yellow'])
plt.axis("equal")
plt.title("Porcentajes de recuperados contra decesos")
plt.show()
```
