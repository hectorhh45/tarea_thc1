```python
# Función para convertir unidades de longitud
def convertir_longitud():
    print("Seleccione una opción de conversión de longitud:")
    print("1. SI a Sistema Inglés")
    print("2. Sistema Inglés a SI")
    opcion = input("Ingrese su opción: ")

    try:
        opcion = int(opcion)
        if opcion == 1:
            print("Seleccione una opción de conversión:")
            print("1. Metros a Pies")
            print("2. Kilómetros a Millas")
            conversion_opcion = input("Ingrese su opción de conversión: ")

            if conversion_opcion == "1":
                metros = float(input("Ingrese la longitud en metros: "))
                pies = metros * 3.281
                print(f"{metros} metros es igual a {pies} pies.")
            elif conversion_opcion == "2":
                kilometros = float(input("Ingrese la longitud en kilómetros: "))
                millas = kilometros * 0.621371
                print(f"{kilometros} kilómetros es igual a {millas} millas.")
            else:
                print("Opción no válida.")
        elif opcion == 2:
            print("Seleccione una opción de conversión:")
            print("1. Pies a Metros")
            print("2. Millas a Kilómetros")
            conversion_opcion = input("Ingrese su opción de conversión: ")

            if conversion_opcion == "1":
                pies = float(input("Ingrese la longitud en pies: "))
                metros = pies / 3.281
                print(f"{pies} pies es igual a {metros} metros.")
            elif conversion_opcion == "2":
                millas = float(input("Ingrese la longitud en millas: "))
                kilometros = millas / 0.621371
                print(f"{millas} millas es igual a {kilometros} kilómetros.")
            else:
                print("Opción no válida.")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Por favor, ingrese un número válido.")


# Función para convertir unidades de masa
def convertir_masa():
    print("Seleccione una opción de conversión de masa:")
    print("1. SI a Sistema Inglés")
    print("2. Sistema Inglés a SI")
    opcion = input("Ingrese su opción: ")

    try:
        opcion = int(opcion)
        if opcion == 1:
            print("Seleccione una opción de conversión:")
            print("1. Kilogramos a Libras")
            print("2. Gramos a Onzas")
            conversion_opcion = input("Ingrese su opción de conversión: ")

            if conversion_opcion == "1":
                kilogramos = float(input("Ingrese el peso en kilogramos: "))
                libras = kilogramos * 2.205
                print(f"{kilogramos} kilogramos es igual a {libras} libras.")
            elif conversion_opcion == "2":
                gramos = float(input("Ingrese el peso en gramos: "))
                onzas = gramos * 0.03527396
                print(f"{gramos} gramos es igual a {onzas} onzas.")
            else:
                print("Opción no válida.")
        elif opcion == 2:
            print("Seleccione una opción de conversión:")
            print("1. Libras a Kilogramos")
            print("2. Onzas a Gramos")
            conversion_opcion = input("Ingrese su opción de conversión: ")

            if conversion_opcion == "1":
                libras = float(input("Ingrese el peso en libras: "))
                kilogramos = libras / 2.205
                print(f"{libras} libras es igual a {kilogramos} kilogramos.")
            elif conversion_opcion == "2":
                onzas = float(input("Ingrese el peso en onzas: "))
                gramos = onzas / 0.03527396
                print(f"{onzas} onzas es igual a {gramos} gramos.")
            else:
                print("Opción no válida.")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Por favor, ingrese un número válido.")


# Función para convertir unidades de tiempo
def convertir_tiempo():
    print("Seleccione una opción de conversión de tiempo:")
    print("1. Horas a Segundos")
    print("2. Segundos a Horas")
    print("3. Minutos a Segundos")
    print("4. Segundos a Minutos")
    print("5. Días a Horas")
    print("6. Horas a Milisegundos")
    opcion = input("Ingrese su opción: ")

    try:
        opcion = int(opcion)
        if opcion == 1:
            horas = float(input("Ingrese el tiempo en horas: "))
            segundos = horas * 3600
            print(f"{horas} horas es igual a {segundos} segundos.")
        elif opcion == 2:
            segundos = float(input("Ingrese el tiempo en segundos: "))
            horas = segundos / 3600
            print(f"{segundos} segundos es igual a {horas} horas.")
        elif opcion == 3:
            minutos = float(input("Ingrese el tiempo en minutos: "))
            segundos = minutos * 60
            print(f"{minutos} minutos es igual a {segundos} segundos.")
        elif opcion == 4:
            segundos = float(input("Ingrese el tiempo en segundos: "))
            minutos = segundos / 60
            print(f"{segundos} segundos es igual a {minutos} minutos.")
        elif opcion == 5:
            dias = float(input("Ingrese el tiempo en días: "))
            horas = dias * 24
            print(f"{dias} días es igual a {horas} horas.")
        elif opcion == 6:
            horas = float(input("Ingrese el tiempo en horas: "))
            milisegundos = horas * 3600000
            print(f"{horas} horas es igual a {milisegundos} milisegundos.")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Por favor, ingrese un número válido.")


# Bucle principal del programa
while True:
    print("Menú Principal:")
    print("1. Convertir unidades de longitud")
    print("2. Convertir unidades de masa")
    print("3. Convertir unidades de tiempo")
    print("4. Salir del programa")
    opcion = input("Seleccione una opción: ")

    try:
        opcion = int(opcion)
        if opcion == 1:
            convertir_longitud()
        elif opcion == 2:
            convertir_masa()
        elif opcion == 3:
            convertir_tiempo()
        elif opcion == 4:
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción del menú.")
    except ValueError:
        print("Por favor, ingrese un número válido.")
```
