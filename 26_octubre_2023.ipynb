```python
#graficar sen y cos

x = np.linspace(0, 2 * np.pi, 1000)

y_sen = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_sen, label='seno', color='blue')
plt.plot(x, y_cos, label='coseno', color='green')

plt.title('Gráfico de las funciones seno y coseno')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.grid(color = 'black', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
def graficación(y):
    x = np.linspace(0, 2 * np.pi, 100)  # Rango de 0 a 2*pi
    y_seno = np.sin(x) + y
    y_coseno = np.cos(x) + y

    # Crear el gráfico
    plt.figure(figsize=(14, 5))

    plt.plot(x, y_seno, label=f'Seno (y={y})', color='blue')
    plt.plot(x, y_coseno, label=f'Coseno (y={y})', color='red')

    plt.grid(axis='y', linestyle='--', color = 'green')
    plt.grid(axis='x', linestyle='--', color = 'yellow')

    plt.xlabel('Ángulo (radianes)')
    plt.ylabel('Valor')
    plt.title('Gráfico de Seno y Coseno')
    plt.legend()
    plt.show()
graficación(0.5)
```
