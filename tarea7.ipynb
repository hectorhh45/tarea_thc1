```python
def es_primo(numero):
    if numero <= 1:
        return "Error en el Número"

    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False

    return True

while True:
    try:
        entrada = int(input("Ingrese un número entero mayor que 1: "))
        resultado = es_primo(entrada)

        if resultado == "Error en el Número":
            print(resultado)
        elif resultado:
            print("True: El número es primo.")
        else:
            print("False: El número no es primo.")

        break  # Salir del bucle si la entrada es válida

    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número entero.")
```

```python
import matplotlib.pyplot as plt

# Crear un arreglo de números enteros entre 0 y 150
numeros = [i for i in range(151)]

# Función para calcular el promedio
promedio = lambda nums: sum(nums) / len(nums)

# Función para calcular la varianza
varianza = lambda nums, avg: sum((x - avg) ** 2 for x in nums) / len(nums)

# Función para calcular la desviación estándar
desviacion_estandar = lambda var: var ** 0.5

# Filtrar números pares e impares
numeros_pares = [num for num in numeros if num % 2 == 0]
numeros_impares = [num for num in numeros if num % 2 != 0]

# Filtrar múltiplos de 2, 3 y 5
multiplos_dos = [num for num in numeros if num % 2 == 0]
multiplos_tres = [num for num in numeros if num % 3 == 0]
multiplos_cinco = [num for num in numeros if num % 5 == 0]

# Calcular y mostrar resultados
avg = promedio(numeros)
var = varianza(numeros, avg)
std_dev = desviacion_estandar(var)

print("Promedio:", avg)
print("Varianza:", var)
print("Desviación Estándar:", std_dev)

# Visualizar histograma
plt.hist(numeros, bins=30, color='blue', edgecolor='black')
plt.title('Histograma de Números entre 0 y 150')
plt.xlabel('Número')
plt.ylabel('Frecuencia')
plt.show()

# Calcular y mostrar la suma de pares e impares
suma_pares = sum(numeros_pares)
suma_impares = sum(numeros_impares)
print("Suma de números pares:", suma_pares)
print("Suma de números impares:", suma_impares)

# Calcular y mostrar múltiplos de 2, 3 y 5
print("Múltiplos de 2:", multiplos_dos)
print("Múltiplos de 3:", multiplos_tres)
print("Múltiplos de 5:", multiplos_cinco)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de la solución general
def solucion_general(x, C):
    return (x - 1) + C * np.exp(-x)

# Valores de x
x_values = np.linspace(-2, 5, 400)

# Valores de C para cada caso
C_positivo = 2
C_cero = 0
C_negativo = -2

# Calcular las soluciones para cada valor de C
y_positivo = solucion_general(x_values, C_positivo)
y_cero = solucion_general(x_values, C_cero)
y_negativo = solucion_general(x_values, C_negativo)

# Graficar las soluciones
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_positivo, label='C > 0', color='blue')
plt.plot(x_values, y_cero, label='C = 0', color='green')
plt.plot(x_values, y_negativo, label='C < 0', color='red')

# Etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Familia de Curvas para la Ecuación Diferencial')
plt.legend()

# Señalar puntos específicos
plt.scatter([0], [4], color='black', marker='o', label='Punto Inicial (x=-1, y=5)')
plt.scatter([1], [0], color='black', marker='o')  # Cero para C = 0

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return 3*x + 7*x**2 - 4

# Rango y número de muestras
x_values = np.linspace(-10, 10, 1000)

# Graficar la función con línea continua
plt.plot(x_values, f(x_values), label='3x + 7x^2 - 4', color='blue')

# Graficar las asíntotas horizontales (y = constante)
plt.axhline(0, linestyle='--', color='red', label='Asíntota horizontal en y=0')

# Graficar las asíntotas verticales (x = constante)
plt.axvline(-1/7, linestyle='--', color='green', label='Asíntota vertical en x=-1/7')

# Establecer límites en el eje y
plt.ylim(-100, 100)

# Añadir leyenda
plt.legend()

# Etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la Función con Asíntotas')

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
